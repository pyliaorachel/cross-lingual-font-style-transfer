```python
import fontforge
font_path = '/home/fangyulin/data/Fonts/PingFang TC/PingFang.ttc'
F = fontforge.open(font_path)

```

```python
with open('test_list.out') as f:
    encoding = f.readlines()
encoding = [x.split('.')[0] for x in encoding]
encoding = encoding[0:-1]
print(encoding[0:-1])
```

```python
lst = []
for img_encoding in encoding:
    y = F[int(img_encoding)]
    lst.append((img_encoding, y.unicode))
print(lst)  
```

```python
with open('cus2uni_encoding.txt', 'w') as fp:
    fp.write('\n'.join('{} {}'.format(x[0],x[1]) for x in lst))
```
